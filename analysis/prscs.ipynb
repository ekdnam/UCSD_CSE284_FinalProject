```python
df = bimbam_df
minor_allele_sum = df.iloc[:, 3:].sum(axis=1)

# Calculate the total number of alleles (2 alleles per individual)
num_alleles = 2 * df.shape[1] - 6  # Subtract 6 to account for the first 3 columns and adjust for 0-based indexing

# Calculate the MAF for each SNP
maf = minor_allele_sum / num_alleles

# Append the MAF as a new column to the DataFrame
df['MAF'] = maf

# Display the DataFrame with the new MAF column
df[['MAF']] == 0
```
