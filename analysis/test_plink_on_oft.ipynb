```python
phen = pd.read_csv('./data/oft/phenotypes.csv')
phen = phen[['rfid', columns[0]]]
```

```python
import scipy.stats
```

```python
pval = 0.1
phen = phen.dropna()
d = pd.merge(prs, phen, on=["IID"])
curr_r2 = scipy.stats.pearsonr(d["phen"],d["SCORE"])[0]**2
print("pval=%s, R2=%s"%(pval, curr_r2))
```

```python
filenames = [
    'bodylength_w_tail', 
    'parafat', 
    'bmi_bodylength_wo_tail',
    'bodyweight',
    'retrofat',
    'bmi_bodylength_w_tail',      
    'epifat',
    'bodylength_wo_tail',
    'fasting_glucose'
]
```

```python
out_dir = 'out'
filename = 'bmi_bodylength_wo_tail'
res = {}

for filename in filenames:
    res[filename] = {}
    print(f'\n\n\nGoing thru {filename}\n\n\n')
    for name in columns[1:]:
        res[filename][name] = {}
        print(f'Processing {name}')
        max_p = None
        max_r2 = None
        input_file_template = f'./{out_dir}/{filename}/{filename}'
        for pval in ["0.000001", "0.00001", "0.0001", "0.001", "0.05", "0.1"]:
            prs = pd.read_csv(f"{input_file_template}.{pval}.profile", delim_whitespace=True)
            phen = pd.read_csv('./data/oft/phenotypes.csv')
            phen = phen[['rfid', name]]
            phen = phen.rename(columns={'rfid': 'IID', name: 'phen'})
            phen = phen.dropna()
            d = pd.merge(prs, phen, on=["IID"])
            curr_r2 = scipy.stats.pearsonr(d["phen"],d["SCORE"])[0]**2
            res[filename][name][pval] = curr_r2
#             print("pval=%s, R2=%s"%(pval, curr_r2))
#         print('\n')
```

```python
curr = pd.DataFrame.from_dict(res)
```

```python
max_r2_bmi_oft = {}
final = {}
currTDict = curr.T.to_dict()
for key in currTDict:
    max_r2_global, max_p_global, phenotype = None, None, None
    final[key] = {}
    for phen in currTDict[key]:
        max_r2, max_p = None, None
        for pval in currTDict[key][phen]:
            curr_r2 = currTDict[key][phen][pval]
            if max_r2 is None or max_r2 < curr_r2:
                max_r2 = curr_r2
                max_p = pval
        final[key][phen] = [max_p, max_r2]
        if max_r2_global is None or max_r2_global < max_r2:
            max_r2_global = max_r2
            max_p_global = max_p
            phenotype = phen
    final[key] = [max_r2, max_p, phenotype]
```
